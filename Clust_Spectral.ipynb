```python
# импорт
import numpy as np
from sklearn.datasets import make_blobs, make_circles, make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# создание данных
X_blob, y_blob = make_blobs(n_samples=300,      # кол-во точек
                            n_features=2,       # кол-во фичей (измерений)
                            centers=3,          # кол-во кластеров
                            cluster_std=2.5,    # разброс точек от центра
                            random_state=42)
```

```python
# стандартизация
scaler = StandardScaler()
scaler.fit(X_blob)
# трансформация
X_blob_std = scaler.transform(X_blob)
```

```python
# печать исходных данных
plt.scatter(X_blob_std[:, 0], X_blob_std[:, 1], c=y_blob);
```

```python
# списки параметров
algorithms = ['nearest_neighbors', 'rbf']
clusters =   [1, 2, 3, 4]
```

```python
# обучение + печать
for i, alg in enumerate(algorithms):               # nn & rbf
  plt.figure(figsize=(15, 3))
  for j, clust in enumerate(clusters):             # 1, 2, 3, 4
    plt.subplot(1, 4, j+1)
    # экземпляр класса
    cluster = SpectralClustering(n_clusters=clust, # кол-во кластеров
                                 affinity=alg,     # алгоритм из списка
                                 random_state=42)

    # предсказание метки кластера - нет отдельного метода predict
    y_blob_pred = cluster.fit_predict(X_blob_std)
    # печать
    plt.scatter(X_blob_std[:, 0], X_blob_std[:, 1], c=y_blob_pred)
    plt.xticks([])
    plt.yticks([])
    plt.title(alg + ' ' + str(clust) + ' clusters')
  plt.show()
```
